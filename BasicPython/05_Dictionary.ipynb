```python
students={"student1":{"name":"Pratik","age":"27"}, "student2":{"name":"Max","age":"27"}}
```

```python
students.items()
```

```python
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary comprehension
squareTable={x:x**2 for x in range(5)}
print(squareTable)
```

```python
##conditional dictionary comprehension

even_Square_table={x:x**2 for x in range(10) if x%2==0}

print(even_Square_table)
```

```python
numbers=[1,2,2,3,3,3,4,4,4,4]

frequncy={}

for number in numbers:
    if number in frequncy:
        frequncy[number]+=1
    else:
        frequncy[number]=1

print(frequncy)            
```

```python
## merge dictionary
dic_1={"a":1,"b":2}
dic_2={"b":3,"c":4}

merged_dic={**dic_1,**dic_2}

print(merged_dic)
```
