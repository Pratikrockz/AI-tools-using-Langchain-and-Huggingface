```python
count=0;
while count<5:
    print(count)
    count=count+1
```

```python
## Break
for i in range(10):
    if i ==5:
        break

    print(i)
```

```python
## continue
for i in range(10):
    if i%2==0:
        continue ## skipping the number where even number is present

    print(i)
```
