```python
## find palindrome

def is_palindrome(text):
    text=text.lower().replace(" ","")
    return text == text[::-1]




is_palindrome("a man a plan a canal panama")
```

```python
## factorial

def factorial(n):
    if n==0:
        return 1
    else:
        return n* factorial(n-1)
    

print(factorial(5))    
```

```python
## read a file and count the frequency of words


def find_frequency(file_path):
    count={}
    with open(file_path,'r') as file:
        for line in file:
            words= line.split()
            for word in words:
                word =word.lower().strip('')
                print(word)
                count[word]=count.get(word,0)+1
                
    return count
             
            


print(find_frequency('sample.txt'))
```

```python
## validate an email
import re
def validate_email(mail):
    pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'

    if re.match(pattern, mail):
        print("Email is valid")

    else:
        print("Email is invalid")


validate_email("abc@google.com")  
validate_email("abc@.com")       
```
